```python
import matplotlib.pyplot as plt
import numpy as np
```

###Signal vo = 5Sin(100πt + π/2) as well its 2ms delayed reverberated tone both by sampled by vs = VsSin(1000πt)  

```python
Ao = 5 #Amplitude
fo = 50 #Orginal signals frequency 
To = 1/fo #Orginal signals period

fs = 500 #Sampling frequency
Ts = 1/fs #Sampling perdiod

cycles = 4
t=np.arange(0, cycles*To, Ts)
x= Ao*np.sin(2*np.pi*fo*t)

###Convert time delay in milliseconds to phase shift in degrees
time_delay = 0.002 #2 milliseconds
phase_shift = 360 * (time_delay / To) ###In degrees

delayed_x= Ao*np.sin(2*np.pi*fo*t - phase_shift) #Reverberated tone
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(17,5))

ax1.plot(x)
ax2.plot(delayed_x)
ax3.stem(x)
ax4.stem(delayed_x)

ax1.set_title('Original signal')
ax1.set_xlabel("time")
ax1.set_ylabel("Amplitude")

ax2.set_title('Echoed signal')
ax2.set_xlabel("time")
ax2.set_ylabel("Amplitude")

ax3.set_title('Original sampled signal')
ax3.set_xlabel("samples")
ax3.set_ylabel("Amplitude")

ax4.set_title('Echoed sampled signal')
ax4.set_xlabel("samples")
ax4.set_ylabel("Amplitude")

fig.tight_layout(pad=2.0) ## Create creating space between plots
```
